### check whether datset is imbalenced or not 

here datset is imbalanced so we can perform smote techniques


```python
df.quality.value_counts().plot(kind="bar")
```

### seeing distribution for each column 


```python
fig, axes = plt.subplots(3, 4, figsize=(20, 15))
axes = axes.ravel()

for i, column in enumerate(df.columns):
    sns.histplot(df[column], kde=True, ax=axes[i], alpha=0.7)
    axes[i].set_title(f'Distribution of {column}', fontsize=12)
    axes[i].set_xlabel(column)
    axes[i].set_ylabel('Frequency')

plt.suptitle('Distribution of All Wine Features', fontsize=16)
plt.tight_layout()
plt.show()
```
