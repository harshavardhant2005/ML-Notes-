### how to plot single line graph


```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [10,20,35,48,50]
plt.plot(x,y)
plt.show()
```

```python
x = [1,2,3,4,5]
y = [10,20,35,48,50]
plt.plot(x,y)
plt.xlabel("X - axis")
plt.ylabel("Y - axis")
plt.title("New graph")
plt.show()
```

```python
x = [1,2,3,4,5]
y = [10,20,35,48,50]
plt.plot(x, y, marker='o', linestyle='--', color='b',linewidth =3,markersize=5,markerfacecolor="red")
plt.xlabel("X - axis")
plt.ylabel("Y - axis")
plt.title("New graph")
plt.show()
```

```python
plt.plot(x, y, marker='.', markerfacecolor='red',linewidth=1,markersize=8)
```

### to plot single graph in divide region


```python
x = [1,2,3,4,5]
y = [10,20,35,48,50]
z = [1,5,9,65,23]
```

```python
plt.figure(figsize=(20,10))
plt.subplot(2,2,1)
plt.plot(x,y)
plt.xlabel("X - AXIS")
plt.xlabel("Y - AXIS")
plt.title("plot1")

plt.subplot(2,2,4)
plt.plot(y,z)
plt.xlabel("X - AXIS")
plt.xlabel("Y - AXIS")
plt.title("plot2")

plt.subplot(2,2,3)
plt.plot(x,z)
plt.xlabel("X - AXIS")
plt.xlabel("Y - AXIS")
plt.title("plot2")

plt.subplot(2,2,2)
plt.plot(y,z)
plt.xlabel("X - AXIS")
plt.xlabel("Y - AXIS")
plt.title("plot2")
```

### bar chart

```python
categories = ["A","B","C","D"]
values =[10,20,3,43]
plt.bar(categories,values,color="blue",edgecolor="black")
```

### histogram
--- this chart is used seperate various disturbution of data set in bins..


```python
data =[1,1,1,2,2,2,2,2,5,2,3,3,4,5,5,3,4,5,3,3,2,2]
plt.hist(data,bins=5,edgecolor="black",color="red")
```

### scatter chart
-- this chart used to represnt various data points

```python
x = [1,2,3,4,2]
y = [10,20,35,48,50]

plt.scatter(x,y,marker="x",color="red")
plt.show()
```

### pie chart

```python
labels =["a","b","c","d"]
sizes = [10,20,50,20]
colour = ["violet","red","blue","green"]
explode = (0, 0.1, 0, 0)
plt.pie(sizes,labels=labels,colors=colour,explode=explode,autopct="%1.1f%%")
```

```python

plt.figure(figsize=(20,15))
plt.subplot(3,2,1)
sizes = [10,20,50,20]
colour = ["violet","red","blue","green"]
explode = (0, 0.1, 0, 0)
plt.title("pie chart")
plt.pie(sizes,labels=labels,colors=colour,explode=explode,autopct="%1.1f%%")


plt.subplot(3,2,2)
x = [1,2,3,4,5]
y = [10,20,35,48,50]
plt.plot(x, y, marker='o', linestyle='--', color='b',linewidth =3,markersize=5,markerfacecolor="red")
plt.xlabel("X - axis")
plt.ylabel("Y - axis")
plt.title("New graph")

plt.subplot(3,2,3)
categories = ["A","B","C","D"]
values =[10,20,3,43]
plt.bar(categories,values,color="blue",edgecolor="black")


plt.subplot(3,2,4)
x = [1,2,3,4,2]
y = [10,20,35,48,50]

plt.scatter(x,y,marker="x",color="red")
plt.show()
```
