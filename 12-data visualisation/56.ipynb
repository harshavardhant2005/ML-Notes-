### group by and data aggerate func


```python
grouped_mean = df.groupby("Product")["Sales"].mean()
```

```python
print(grouped_mean)
```

```python
grouped_means = df.groupby(["Product","Region"])["Sales"].sum()
grouped_means
```

```python
grouped = df.groupby(["Product","Region"])["Sales"].agg(["mean","sum","count"])
grouped
```

### merging two tables

```python
df1= pd.DataFrame({"key":["A","B","C","Y"],"values":["1","2","3","4"]})
df1
```

```python
df2 = pd.DataFrame({"key":["A","E","Y","X"],"values":["100","95","58","56"]})
df2
```

```python
merged=pd.merge(df1,df2,on="key",how="right")
print(merged)
```
