### importing df


```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns 
```

### loading dataset 

```python
df = pd.read_excel('flight_price.xlsx')
df.head()
```

```python
df.info()
```

***

### converting dept time into numeric features


```python
df["Dep_Hours"] = df["Dep_Time"].str.split(":").str[0].astype(int)
df["Dep_Minutes"] = df["Dep_Time"].str.split(":").str[0].astype(int)
df.drop(columns="Dep_Time",axis=1,inplace=True)
```

```python
df.info()
```

***

### converting date of journey into seperate columns 

```python
df["Date"] = df["Date_of_Journey"].str.split("/").str[0].astype(int)
df["Month"] = df["Date_of_Journey"].str.split("/").str[1].astype(int)
df["Year"] = df["Date_of_Journey"].str.split("/").str[2].astype(int)

```

```python
df.drop(columns="Date_of_Journey",inplace=True,axis=1)
```

```python
df.head()
```

***

### convert arrival time 

```python
df["Arrival_Time"] = df["Arrival_Time"].str.split(" ").str[0]
```

```python
df["Arrival_Hour"] = df["Arrival_Time"].str.split(":").str[0].astype(int)
df["Arrival_Minutes"] = df["Arrival_Time"].str.split(":").str[1].astype(int)
```

```python
df.drop("Arrival_Time",inplace=True,axis=1)
```

```python
df.head()
```

***

### converting Totsl stops into encoding 

```python
df["Total_Stops"].unique()
```

```python
encoded = {"1 stop": 1,"2 stops":2,"3 stops":3,"4 stops":4,"non-stop":1,np.nan:"0"}

df["Total_Stop"] = df["Total_Stops"].map(encoded).astype(int)

```

```python
df.drop("Total_Stops",inplace=True,axis=1)
```

```python
df.head(5)
```

***

### converting duration hours and minutes into duration with respect to minutes

```python
hours = df["Duration"].str.split(" ").str[0].str.split("h").str[0].str.split("m").str[0].astype(int)
minutes = df["Duration"].str.split(" ").str[1].str.split("m").str[0]
minutes = minutes.fillna(0).astype(int)
df["Duration_minutes"] = hours*60+minutes
```

```python
df.drop("Duration",axis=1,inplace=True)
```

***

### drop Route feature because we already know number of stop it contain 

```python
df.drop("Route",inplace=True,axis=1)
```

***

### label encoding for the airline name and other features remenaing

```python
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder()
encoded = encoder.fit_transform(df[["Airline","Source","Destination"]]).toarray()
encoded_df = pd.DataFrame(encoded,columns=encoder.get_feature_names_out()).astype(int)
```

```python
df=pd.concat([df,encoded_df],axis=1)
df.drop(["Airline","Source","Destination"],axis=1,inplace=True)
```
